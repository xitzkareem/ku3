```python

from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
browser = webdriver.Firefox() 
```

```python
url2 = 'https://finance.yahoo.com/'
browser.get(url2)
print(f"Successfully visited: {browser.title} ")
```

```python
button = browser.find_element(By.NAME, 'agree')
button.click()
```

```python
headlines = browser.find_elements(By.CLASS_NAME, 'yf-1u32w3i')
print(f"Found {len(headlines)} headlines.")
# Print the first 5 headlines found
for i, headline in enumerate(headlines[:5]):
    print(f"{i+1}: {headline.text}")
```

```python
headlines = browser.find_elements(By.CLASS_NAME, 'clamp')
print(f"Found {len(headlines)} headlines.")
# Print the first 5 headlines found
for i, headline in enumerate(headlines[:5]):
    print(f"{i+1}: {headline.text}")
```

```python
headlines = browser.find_elements(By.CSS_SELECTOR, 'h3.tw-line-clamp-none')
print(f"Found {len(headlines)} headlines.")
# Print the first 5 headlines found
for i, headline in enumerate(headlines[:5]):
    print(f"{i+1}: {headline.text}")
```

```python
headlines = browser.find_elements(By.TAG_NAME, 'h3')
print(f"Found {len(headlines)} headlines.")
# Print the first 5 headlines found
for i, headline in enumerate(headlines[:5]):
    print(f"{i+1}: {headline.text}")
```
